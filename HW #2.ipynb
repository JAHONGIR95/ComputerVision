```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
image = cv2.imread('global.jpg')
img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.title("Real Image")
plt.show()
```

```python
#Implement Kernel Convolution
#Identity matrix

identity = np.array(([0,0,0],[0,1,0],[0,0,0]),np.float32)
print("Identity Matrix")
print(identity)
identity_image = cv2.filter2D(img,-1,identity)
plt.imshow(identity_image)
plt.axis('off')
plt.title("Identity Operation on Image")
plt.show()
```

```python
#Edge Detection on Image
edge1 = np.array(([1,0,-1],[0,0,0],[-1,0,1]),np.float32)
edge2 = np.array(([0,1,0],[1,-4,1],[0,1,0]),np.float32)
edge3 = np.array(([-1,-1,-1],[-1,8,-1],[-1,-1,-1]),np.float32)
edge1_image = cv2.filter2D(img,-1,edge1)
edge2_image = cv2.filter2D(img,-1,edge2)
edge3_image = cv2.filter2D(img,-1,edge3)
fig, axs = subplots(1, 3, figsize=(16, 16))
axs[0].imshow(edge1_image)
axs[0].axis('off')
axs[1].imshow(edge2_image)
axs[1].axis('off')
axs[2].imshow(edge3_image)
axs[2].axis('off')
show()
```

```python
#Sharpen Image
sharpen = np.array(([0,-1,0],[-1,5,-1],[0,-1,0]),np.float32)
sharpen_image = cv2.filter2D(img,-1,sharpen)
plt.imshow(sharpen_image)
plt.rcParams['figure.figsize'] =(16,16)
plt.axis('off')
plt.title("Sharpen Image")
plt.show()
```

```python
#Box Blur
box_blur = np.array(([0,-1,0],[-1,5,-1],[0,-1,0]),np.float32) / 9
boxblur_image = cv2.filter2D(img,-1,box_blur)
plt.imshow(boxblur_image)
plt.axis('off')
plt.title("Box Blur Image")
plt.show()
```

```python
#Gaussian Blur
gaussian_blur = np.array(([1,2,1],[2,4,2],[1,2,1]),np.float32) / 16
gaussianblur_image = cv2.filter2D(img,-1,gaussian_blur)
plt.imshow(gaussianblur_image)
plt.axis('off')
plt.title("Gaussian Blur Image")
plt.show()
```

```python
#Sobel Operator
sobel = np.array(([-1,0,1],[-2,0,2],[-1,2,1]),np.float32)
sobel_image = cv2.filter2D(img,-1,sobel)
plt.imshow(sobel_image)
plt.axis('off')
plt.title("Sobel Operator Image")
plt.show()
```
